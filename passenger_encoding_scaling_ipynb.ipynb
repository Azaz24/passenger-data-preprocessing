
## Title: Passenger Data Encoding & Feature Scaling
### Objective: Prepare dataset for ML model using encoding and scaling techniques.



```python
# Import & create Dataset

import pandas as pd

data = {
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Female'],
    'City': ['Mumbai', 'Delhi', 'Chennai', 'Mumbai', 'Delhi'],
    'Size': ['Small', 'Large', 'Medium', 'Large', 'Small'],
    'Age': [25, 45, 32, 60, 28],
    'Fare': [15, 300, 85, 450, 20]
}

df = pd.DataFrame(data)
df
```

# TASK 1: CATEGORICAL ENCODING

```python
# 2.1 Gender → Label Encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
```

Conclusion: The ML data has been converted into numeric format.

```python
# 2.2 City → One Hot Encoding

df = pd.get_dummies(df, columns=['City'], drop_first=True)
```

Conclusion: The model does not get confused and there is no incorrect priority order.

```python
# 2.3 Size → Ordered Label Encoding

size_mapping = {'Small': 0, 'Medium': 1, 'Large': 2}
df['Size'] = df['Size'].map(size_mapping)
```

Conclusion: The order is preserved and the ML model correctly understands the size hierarchy.

# TASK 2: FEATURE SCALING

###Purpose: Scaling ensures fair comparison between features

```python
# Apply Scaling (StandardScaler)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
df

# Reason for using StandardScaler:
# - Fare column contains outlier (450)
# - StandardScaler normalizes data (mean = 0, std = 1)
# - Helps ML model treat Age and Fare equally
```
